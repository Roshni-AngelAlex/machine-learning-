```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.show()
# SCENARIO 1: E-Commerce Sales Data
print("NAME:ROSHNI ANGEL A,ROLLNO:24BAD101")
# 1. Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# 2. Load the dataset
# (Make sure the CSV file is in the same folder as this notebook)
df = pd.read_csv(r"C:\Users\Alex Roshni Angel\Downloads\ML\archive.zip", encoding="ISO-8859-1")

# 3. Inspect the dataset
print("----- HEAD -----")
print(df.head())

print("\n----- TAIL -----")
print(df.tail())

print("\n----- INFO -----")
print(df.info())

print("\n----- DESCRIBE -----")
print(df.describe())

# 4. Check for missing values
print("\n----- MISSING VALUES -----")
print(df.isnull().sum())

# 5. Data preprocessing
# Convert InvoiceDate to datetime format
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])

# 6. Visualize sales per product (Bar Chart)
product_sales = df.groupby("Description")["Quantity"].sum().sort_values(ascending=False).head(10)

plt.figure()
product_sales.plot(kind="bar")
plt.xlabel("Product")
plt.ylabel("Total Quantity Sold")
plt.title("Top 10 Products by Sales")
plt.xticks(rotation=90)
plt.show()

# 7. Visualize sales trend over time (Line Chart)
daily_sales = df.groupby(df["InvoiceDate"].dt.date)["Quantity"].sum()

plt.figure()
plt.plot(daily_sales)
plt.xlabel("Date")
plt.ylabel("Total Sales Quantity")
plt.title("Daily Sales Trend")
plt.show()
```

```python
# SCENARIO 2: Hospital Patient Records
# OBJECTIVE: Identify missing health metrics and visualize distributions
print("NAME:ROSHNI ANGEL A,ROLLNO:24BAD101")

# 1. Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Load the dataset
# Ensure diabetes.csv is in the same folder
df = pd.read_csv(r"C:\Users\Alex Roshni Angel\Downloads\ML\archive (1).zip")

# 3. Explore the dataset structure
print("----- DATASET HEAD -----")
print(df.head())

print("\n----- DATASET INFO -----")
print(df.info())

print("\n----- DATASET DESCRIPTION -----")
print(df.describe())

# 4. Check for missing values
print("\n----- MISSING VALUES -----")
print(df.isnull().sum())

# 5. Identify zero values (treated as missing in medical data)
zero_columns = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]
print("\nZero Values Count:")
print((df[zero_columns] == 0).sum())

# 6. Visualization: Glucose level histogram
plt.figure()
plt.hist(df["Glucose"], bins=20)
plt.xlabel("Glucose Level")
plt.ylabel("Frequency")
plt.title("Distribution of Glucose Levels")
plt.show()

# 7. Visualization: Age distribution histogram
plt.figure()
plt.hist(df["Age"], bins=20)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution")
plt.show()
# 8. Boxplot for glucose levels
plt.figure()
sns.boxplot(x=df["Glucose"])
plt.title("Boxplot of Glucose Levels")
plt.show()

# 9. Boxplot for BMI
plt.figure()
sns.boxplot(x=df["BMI"])
plt.title("Boxplot of BMI")
plt.show()

# 10. Correlation analysis
correlation_matrix = df.corr()

# 11. Heatmap visualization
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Health Metrics")
plt.show()

# 12. Observations
print("\nObservations:")
print("Glucose, BMI, and Age show noticeable correlation with diabetes outcome.")
```

```python
# SCENARIO 3: Housing Dataset
# OBJECTIVE: Analyze missing values and relationships between house attributes
print("NAME:ROSHNI ANGEL A,ROLLNO:24BAD101")
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv(r"C:\Users\Alex Roshni Angel\Downloads\ML\archive (2).zip")

# Inspect dataset (for record)
print(df.head())
print(df.info())
print(df.describe())

# Bedrooms vs Price
plt.figure()
plt.scatter(df["bedrooms"], df["price"])
plt.xlabel("Number of Bedrooms")
plt.ylabel("Price")
plt.title("Bedrooms vs Price")
plt.show()

# Area vs Price
plt.figure()
plt.scatter(df["area"], df["price"])
plt.xlabel("Area (sq ft)")
plt.ylabel("Price")
plt.title("Area vs Price")
plt.show()

# Bathrooms vs Price
plt.figure()
plt.scatter(df["bathrooms"], df["price"])
plt.xlabel("Number of Bathrooms")
plt.ylabel("Price")
plt.title("Bathrooms vs Price")
plt.show()

# Correlation Heatmap (Numerical Features Only)
correlation_matrix = df.select_dtypes(include='number').corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Housing Features")
plt.show()

# Interpretation
print("\nInterpretation:")
print("Area, bedrooms, and bathrooms show positive correlation with house price.")
print("Area has the strongest influence on price prediction.")
```

```python
# SCENARIO 4: Banking Customer Data
# OBJECTIVE: Analyze customer demographics and spending patterns
print("NAME:ROSHNI ANGEL A,ROLLNO:24BAD101")
# 1. Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Load the dataset
# File name: marketing_campaign.csv
df = pd.read_csv(r"C:\Users\Alex Roshni Angel\Downloads\ML\archive (3).zip", sep="\t")

# 3. Examine structure
print("----- DATASET HEAD -----")
print(df.head())

print("\n----- DATASET INFO -----")
print(df.info())

print("\n----- MISSING VALUES -----")
print(df.isnull().sum())

# 4. Create Age column
df["Age"] = 2024 - df["Year_Birth"]

# Bar plot: Age distribution (Top 10 ages)
plt.figure()
df["Age"].value_counts().sort_index().head(10).plot(kind="bar")
plt.xlabel("Age")
plt.ylabel("Number of Customers")
plt.title("Age Distribution (Top Ages)")
plt.show()

# Box plot: Age distribution
plt.figure()
sns.boxplot(x=df["Age"])
plt.title("Age Distribution (Boxplot)")
plt.show()
# Box plot: Income distribution
plt.figure()
sns.boxplot(x=df["Income"])
plt.title("Income Distribution")
plt.show()

# Bar plot: Marital Status
plt.figure()
df["Marital_Status"].value_counts().plot(kind="bar")
plt.xlabel("Marital Status")
plt.ylabel("Count")
plt.title("Customer Marital Status Distribution")
plt.show()

# Bar plot: Spending on product categories
spending_cols = ["MntWines", "MntFruits", "MntMeatProducts",
                 "MntFishProducts", "MntSweetProducts"]

plt.figure()
df[spending_cols].sum().plot(kind="bar")
plt.xlabel("Product Category")
plt.ylabel("Total Spending")
plt.title("Customer Spending Pattern")
plt.show()

# Box plot: Total spending distribution
df["Total_Spending"] = df[spending_cols].sum(axis=1)

plt.figure()
sns.boxplot(x=df["Total_Spending"])
plt.title("Total Customer Spending Distribution")
plt.show()

# 5. Insights Summary
print("\nInsights:")
print("Most customers belong to middle-age groups.")
print("Income shows wide variation with outliers.")
print("Wine and meat products contribute the highest spending.")
print("Customer spending varies significantly across individuals.")
```
